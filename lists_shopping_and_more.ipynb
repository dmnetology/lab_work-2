Задание: написать программу, которая поможет пользователям определить, сколько денег они могут сэкономить на покупках в разных магазинах.

Условия:

Пользователь должен ввести список товаров, которые он собирается купить, и цены на эти товары в разных магазинах.

Затем программа должна вывести список магазинов и общую стоимость покупок в каждом магазине, а также сообщить, в каком магазине пользователь может сэкономить больше всего денег.

Список товаров и магазинов определяется самостоятельно.

```python
# На вход функции передается список покупок и данные ценам определенного магазина
# Если все товары из списка покупок доступны, то возвращаем сумму, иначе "недоступно"
def check_sum(shopping_list, grocery):
    total = 0
    for el in shopping_list:
        if el[0] in grocery:
            total += int(el[1]) * grocery[el[0]]
        else:
            return -1
    return total
```

```python
# Считаем сумму в каждом магазине и заносим в список
def sum_grocery(shopping_list, groceries):
    lst = []
    for el in groceries:
        sum_gr = check_sum(shopping_list, el[1])
        if sum_gr > 0:
            lst.append([el[0], sum_gr])
    return lst
```

```python
# Доступные магазины.
names = ['Пятерочка', 'Лента', 'Метро', 'Экзотика', 'За углом']
# Данные о ценах на товары в магазинах.
prices = [{'хлеб' : 54, 'молоко' : 77, 'сметана' : 80, 'яблоки' : 131, 'кефир' : 94, 'торт' : 512},\
          {'хлеб' : 62, 'молоко' : 76, 'сметана' : 77, 'яблоки' : 160, 'торт' : 480},\
          {'хлеб' : 59, 'молоко' : 81, 'сметана' : 72, 'яблоки' : 120, 'кефир' : 90, 'торт' : 498},\
          {'бананы' : 88, 'ананас' : 300, 'киви' : 110},\
          {'хлеб' : 40, 'молоко' : 65}]
# Объединение данных о магазинах и ценах
groceries = zip(names, prices)
# Получим общий словарь со всеми ключами
product = {}
for d in prices:
    product.update(d)
# Ввод списка продуктов, которые хотим приобрести, а также его количества
shopping_list = []
print("Продукты, доступные к покупке (некоторые продукты могут быть недоступны в отдельном магазине):")
print(*product.keys(), sep=', ')
print('========================================================================')
print("Введите продукт и количество через запятую: . Закончить ввод - любая цифра")
while True:
    product = input().split(sep=',')
    if not product[0].isdigit():
        shopping_list.append(product)
    else:
        break

# Получить список сумм покупок в каждом магазине при условии, что продукты в магазине полностью доступны
# При недоступности какого-либо товара, магазин исключается из списка доступных
lst_sum = sorted(sum_grocery(shopping_list, groceries), key=lambda x: x[1])
for el in lst_sum:
    print(f'В магазине "{el[0]}" покупка Вам обойдется в {el[1]} руб.')
print()    
if len(lst_sum):
    print('Самая большая экономия будет в магазине "{}"'.format(lst_sum[0][0]))
else:
    print('К сожалению, нет магазина, где Вы могли бы купить это все разом')
```
